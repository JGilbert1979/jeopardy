```python
#useful imports for normalization and tokenization
import nltk
from collections import Counter
from nltk.corpus import stopwords
from pprint import pprint
from string import punctuation
import numpy as np
from nltk.probability import FreqDist
```

```python
sw = stopwords.words('english')
```

```python
#cleaning steps
jeopardy = open("JEOPARDY_CSV.csv",'r', encoding = 'UTF-8').read()
jeopardy_clean = [w for w in jeopardy.lower().split(',')]
jeopardy_clean = [w for w in str(jeopardy_clean).split()]
jeopardy_clean = [w for w in jeopardy_clean if w.isalpha() and w not in sw]

```

```python
jeopardy_clean[:200]
```

```python
#tools to measure tken length and distribution
jeopardy_count = Counter(jeopardy_clean)
len(jeopardy_clean)
jeopardy_fd = FreqDist(jeopardy_clean)
jeopardy_fd
```

```python
#descriptive statistics tools
print(f"Jeopardy is {len(jeopardy_clean)} tokens long.")
print(f"Jeopardy has {len(set(jeopardy_clean))} unique tokens.")

print(f"Jeopardy's lexical diversity is {len(set(jeopardy_clean))/len(jeopardy_clean):.3f}.")

# Build a vector of token length
jeopardy_token_len = [len(w) for w in jeopardy_clean]

print(f"Jeopardy's average token length is {np.mean(jeopardy_token_len):.2f}.")

pprint(sorted(Counter(jeopardy_token_len).items()))

print("")
print("All statistics are calculated after normalization and tokenization.")
```
